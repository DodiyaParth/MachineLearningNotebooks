```python
#concat
us_weather=pd.DataFrame({
    'city':["new York","chicago","orlando"],
    'temprature':[50,60,45],
    'humidity':[50,55,20]
})
us_weather
```

```python
india_weather=pd.DataFrame({
    'city':["mumbai","delhi","banglore"],
    'temprature':[50,60,45],
    'humidity':[50,55,20]
})
india_weather
```

```python
df=pd.concat([india_weather,us_weather])
df
```

```python
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

##### row wise concatanation

```python
df=pd.concat([india_weather,us_weather],axis=1)
df
```

```python
# MERGE
temprature_df=pd.DataFrame({
    "city":['mumbai','delhi','banglore','hydrabad'],
    'temprature':[32,45,30,40]
})
temprature_df
```

```python
humidity_df=pd.DataFrame({
    "city":['mumbai','delhi','banglore'],
    'humidity':[71,60,20]
})
humidity_df
```

```python
# it will drop hydrabad
df=pd.merge(temprature_df,humidity_df,on='city')
df
```

```python
#to take hydrabad
#outer join
df=pd.merge(temprature_df,humidity_df,on='city',how='outer')
df
```
