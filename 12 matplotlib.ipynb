### ******************************EXPERIMENT***********************

```python
n=150
lst=[i for i in range(n) if i%2!=0]
def prime(a):
    for i in range(2,a):
        if a%i==0:
            return False
    return True
primes=list(filter(prime,lst))
#print(primes)
print(len(primes))
x=[i for i in range(1,len(primes)+1)]
```

```python
plt.plot(x,primes,'g:')
plt.grid()
plt.xlabel("index of prime")
plt.ylabel("prime number")
plt.title("Prime Numbers")
plt.show()
```

```python
import pandas as pd
import numpy as np
p={"index":np.array(x),"prime_Value":np.array(primes)}
df=pd.DataFrame.from_dict(p)
```

```python
df.to_csv("primes.csv")
```
